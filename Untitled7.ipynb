```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import os
import pandas as pd

newdf = pd.read_csv('data.csv')

print(newdf.head())
```

```python
df2=newdf.pivot_table(values='opening_rank', columns=['year','category'],index=['program_name','institute_short'],aggfunc='max')
```

```python
institute=newdf.groupby('institute_type').institute_short.value_counts()
institute=pd.DataFrame(institute)
institute.rename(columns={'institute_short':'no_of_seats'},inplace=True)
institute.reset_index(inplace=True
                     )
```

```python
iit=institute['institute_type']=='IIT'
IIT=institute[iit]
```

```python
plt.title("No of Seats in IIT ")
plt.xticks(rotation=90);
sns.barplot(x="institute_short",y="no_of_seats",data=IIT)
plt.show()
```

```python
NIT=institute[-iit]
```

```python
plt.title("No of Seats in NIT ")
plt.xticks(rotation=90);
sns.barplot(x="institute_short",y="no_of_seats",data=NIT)
plt.show()
```

```python
oran=newdf.groupby('institute_short').opening_rank.max()
oran=pd.DataFrame(oran)
```

```python
oran.reset_index(inplace=True)
oran=oran.head(12)
```

```python
cran=newdf.groupby('institute_short').opening_rank.min()
cran=pd.DataFrame(cran)
cran.rename(columns={"opening_rank":"closing_rank"},inplace=True)
cran.reset_index(inplace=True)
cran=cran.head(12)
```

```python
oran
```

```python
plt.xticks(rotation=90);
plt.title("Opening vs closing ranks in IIT")
sns.lineplot(x="institute_short",y="closing_rank",data=cran)
sns.lineplot(x="institute_short",y="opening_rank",data=oran)
plt.legend(["close_rank","open_rank"])
plt.show()
```

```python
sns.lineplot(x="institute_short",y="opening_rank",data=oran)
plt.xticks(rotation=90)
plt.show()
```

```python
testing=newdf.groupby('quota').year.value_counts()
testing=pd.DataFrame(testing)
testing.rename(columns={"year":"count"},inplace=True)
testing.reset_index(inplace=True)
testing['year'].astype(str)
testing['year'] = testing['year'].map(str)
testing['year']=testing['quota'] + ' ' + testing['year']
```

```python
plt.xticks(rotation=90)
sns.barplot(x="year",y='count',data=testing)
plt.show()
```
